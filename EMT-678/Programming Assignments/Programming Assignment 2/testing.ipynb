```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('data/Class 9 - 12  - Data for Programming - Environmental - vshort.xlsx', header=None)
data.head(10)
```

```python
# every fifth row is a province
provinces = data.iloc[::6, 0].tolist()
measurements = data.iloc[1:5, 0].tolist()
headers = data.iloc[0, 1:].tolist()

# take a frame of data every 5th row
subsets = {}

for i in range(len(provinces)):

    subset = data.iloc[i*6+1:i*6+5, 1:]
    subsets.update({provinces[i]: subset})

print(provinces)
print(measurements)
print(headers)

subsets['Alberta']
```

```python
subsetted_data = np.vstack([subsets[province].values for province in provinces])
```

```python
multi_index = pd.MultiIndex.from_product([provinces, measurements], names=['Province', 'Measurement'])
df = pd.DataFrame(index=multi_index, columns=headers, data=subsetted_data)
df
```
