## Objective
The goal of this activity is to build a simple language model by tokenizing a
text corpus, counting word occurrences, and plotting the frequency distribution
of words.
## Instructions
1. Tokenize the Corpus: Use the following corpus for tokenization.
”The quick brown fox jumps over the lazy dog. The fox was quick to jump.”
Tokenize the text into individual words. Be sure to:
    * Exclude punctuation.

    * Convert all words to lowercase.

2. Build a Frequency Dictionary: After tokenizing, count the occur-
rences of each unique word and create a dictionary that maps each word
to its frequency. The result should look like this:
{“the”: 3, “quick”: 2, “fox”: 2, ...}

3. Convert to a Probability Distribution: Calculate the probability of
each word by dividing its frequency by the total number of words. This
will create a simple unigram language model:
P(word) = Frequency of word / Total number of words

4. Plot the Distribution: Use Python’s matplotlib library to create a bar
plot of the probability distribution. The x-axis should represent words,
and the y-axis should represent their probabilities.

## 1)

```python
def tokenize(corpus: str):
    corpus = corpus.lower()
    corpus = corpus.replace('.','').replace(',','')
    return corpus.lower().split()
```

```python
corpus = 'the quick brown fox jumps over the lazy dog. The fox was quick to jump.'
tokens = tokenize(corpus)

print(tokens)
```

## 2)

```python
freq_dist = {}
for unique_token in set(tokens):
    freq_dist[unique_token] = tokens.count(unique_token)

for freq_entry in sorted(freq_dist.items(), key=lambda x: x[1], reverse=True):
    print(freq_entry)
```

## 3)

```python
num_tokens = len(tokens)
prob_dist = {k: v/num_tokens for k, v in freq_dist.items()}
for unique_token, prob in sorted(prob_dist.items(), key=lambda x: x[1], reverse=True):
    print(f'{unique_token}: {prob:.1%}')
```

## 4)

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 1, figsize=(10, 5))

axs.bar(prob_dist.keys(), prob_dist.values())
axs.grid(axis='y')

axs.set_xlabel('Token')
axs.set_ylabel('Probability of Token')
axs.set_title('Probability Distribution of Tokens')

plt.show()
```
